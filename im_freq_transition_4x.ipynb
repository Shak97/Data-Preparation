```python
import os
import numpy as np
import scipy.misc
import imageio
import nibabel as nib
import matplotlib.pyplot as plt
%matplotlib inline
import imageio
```

```python
dirpath = 'recons_data\\dataset\\'
```

```python
dir_names = os.listdir(dirpath)
filename = dirpath + dir_names[1] + '\\ses-BL\\anat\\' + dir_names[1] + '_ses-BL_T1w.nii.gz'
images = nib.load(filename)
data = images.get_fdata(caching = 'fill')
data = data.astype(np.uint16)
```

```python
data.shape
```

```python
gaussian_mat = np.zeros(shape = (256,256))
for i in gaussian:
    for j in range(256):
        gaussian_mat[i,j] = 1
```

```python
plt.imshow(gaussian_mat, 'gray')
imageio.imsave('gaussian_mat4x.png', gaussian_mat)
```

```python
plt.imshow(data[:,70,:], 'gray')
imageio.imsave('gaussian4ximdomain.png',data[:,70,:])
```

```python
fft_shift_data = np.fft.fftshift(np.fft.fft2(data[:,70,:]))
```

```python
plt.imshow(np.abs(fft_shift_data), 'gray')
imageio.imsave('gaussian4xfreqdomain.png', np.abs(fft_shift_data))
```

```python
plt.imshow(np.abs(np.multiply(fft_shift_data, gaussian_mat)), 'gray')
imageio.imsave('gaussian4xsubsampled.png', np.abs(np.multiply(fft_shift_data, gaussian_mat)))
```

```python
res = np.abs(np.real(np.fft.ifft2(np.fft.ifftshift(np.multiply(gaussian_mat, fft_shift_data)))))
```

```python
plt.imshow(res, 'gray')
imageio.imsave('gaussiansubsampled4ximmdomain.png', res)
```
